```python
import pandas as pd
```

```python
zillow_data1= "zillow_data/zillow_medianhome_matched.csv"
zillow_pd= pd.read_csv(zillow_data1)
zillow_pd.head()
```

```python
census_data1= "census_data/combined_census_cleaned.csv"
census_pd= pd.read_csv(census_data1)
census_pd.head()
```

```python
merge_data= pd.merge(zillow_pd,census_pd, on="Zip")
merge_data
```

```python
merge_data = merge_data.drop(["City_y"], axis=1)
merge_data
```

```python
merge_data= merge_data.rename(columns={"City_x":"City", "Year_x":"Zillow Years", "Year_y":"Census Years"})
merge_data
```
