```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
zillow_data1= "data_sources/zillow_data/zillow_medianhome_matched.csv"
zillow_pd= pd.read_csv(zillow_data1)
zillow_pd.head()
```

```python
census_data1= "data_sources/census_data/combined_census_cleaned.csv"
census_pd= pd.read_csv(census_data1)
census_pd.head()
```

```python
merge_data= pd.merge(zillow_pd,census_pd, on=["Zip","Year"])
merge_data
```

```python
merge_data = merge_data.drop(["City_y"], axis=1)
merge_data
```

```python
merge_data= merge_data.rename(columns={"City_x":"City"})
merge_data
```

```python
Urban= merge_data[merge_data["Type"]=="Urban"]
Rural= merge_data[merge_data["Type"]=="Rural"]
Suburban= merge_data[merge_data["Type"]=="Suburban"]
Suburban
```

```python
Urban_Div= Urban.groupby(["Zip"])["Diversity Index"].mean()
Urban_HV= Urban.groupby(["Zip"])["Value"].mean()

Rural_Div= Rural.groupby(["Zip"])["Diversity Index"].mean()
Rural_HV= Rural.groupby(["Zip"])["Value"].mean()

Suburban_Div= Suburban.groupby(["Zip"])["Diversity Index"].mean()
Suburban_HV= Suburban.groupby(["Zip"])["Value"].mean()

plt.scatter(Urban_Div,Urban_HV, facecolor= "blue", edgecolor= "black", marker="o", label= "Urban", s=3*Urban["Zip"].count())
plt.scatter(Rural_Div,Rural_HV, facecolor= "yellow", edgecolor= "black", marker="o", label= "Rural", s=3*Rural["Zip"].count())
plt.scatter(Suburban_Div, Suburban_HV, facecolor= "red", edgecolor= "black", marker="o", label= "Suburban", s= Suburban["Zip"].count())

plt.xlabel("Average Diversity")
plt.ylabel("Average Home Value")
plt.title("Diversity affect on Home Values")
plt.grid()
# Create a legend
plt.legend(title="City types")
plt.savefig("output_images/scatter_godwin.png")
plt.show()
```

### This graph supports our alternate hypothesis that Rising property values in Austin are correlated with zip code areas becoming less diverse over time no matter the city type. Common logic would indicate that Rural areas would be lower than Urban and Suburban areas no matter where the average diversity scale falls.
